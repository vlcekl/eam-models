```python
import numpy as np
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
def gauss(x, mu, sig):
    return 1/(2*np.pi*sig**2)**0.5*np.exp(-(x-mu)**2/(2*sig**2))
    
```

```python
rr = np.linspace(5, 20, 101)

plt.figure(figsize=(10,8))

plt.plot(rr, [gauss(r, 10, 2) for r in rr])
plt.plot(rr, [gauss(r, 13, 1) for r in rr])
plt.xlabel('x', fontsize=20)
plt.ylabel('p(x)', fontsize=20)
plt.vlines(10, 0, 0.42)
plt.vlines(13, 0, 0.42)
plt.ylim(0,0.42)
```
